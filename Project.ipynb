# Dataframes:


1. **Human Development Index (hdi):** 
    
    Index used to rank countries by level of "human development". It contains three dimensions: *health level*, *educational level* and *living standard*.
    
    - Value Range: 0 to 1 (at higher values greater, human development)
    - Country Range: 187 countries 
    - Year Range: 1990-2015 (27) 
    
    
2. **Population (pop):**
    
    Total population.
    
    - Country Range: 195 countries
    - Year Range: 1800-2100 (prediction) (302)
    
    
3. **Gross National Income per capita (gni):**
    
    Gross National Income divided by midyear population. Data are in constant 2010 U.S. dollars.
   
    - Country Range: 155 countries
    - Year Range: 1975-2018 (45)


4. **Gender Equality Index (gender_equality):**
    
    Contains five dimensions: *power distribution by gender*, *female participation in civil society organizations*, *ratio between female and male mean years of schooling*, *proportion of lower chamber female legislators* and *proportion of women in ministerial level positions*.
    
    - Value Range: 0 to 100 (at higher values, greater gender equality)
    - Country Range: 186 countries
    - Year Range: 1960-2017 (59)


5. **CO2 Emissions (emissions):**
    
    Carbon dioxide emissions from burning of fossil fuels (metric tonnes of CO2 per person).
    
    - Values: higher values are obviously worse
    - Country Range: 192 countries
    - Year Range: 1800-2014 (216)
    
    
6. **Fundamental Rights (fundrights):**

    It includes three subattributes: *acces to justice*,*civil liberties*, and *social rights and equality*
    
    - Value Range: 0 to 100 (at higher values, more fundamental rights are respected by the country)
    - Country Range: 155 countries
    - Year Range: 1975-2018 (45)
    
    
7. **Murders/100k people (murders):**

    Mortality due to interpersonal violence, per 100,000 standard population, age adjusted. This rate is calculated as if all countries had the same age composition of the population.
    
    - Values: higher values are obviously worse
    - Country Range: 103 countries
    - Year Range: 1950-2016 (68)
    

8. **Press Freedom Index (press_freedom):**

    The data is collected through an online questionnaire sent to journalists, media lawyers, researchers and other media specialists selected by Reporters without Borders (RSF).

    - Values: 0 to 100 (at higher values, the country has a freer press) 
    - Country Range: 180 countries
    - Year Range: 2001-2019 (18)

#Â Data cleaning and manipulation


Once we've seen the dataframes, we want to clean and transform them to obtain a single dataframe with all indicators by year and country.

```python
#First, we'll get the countries all of them share:
countries_in_common = list(set.intersection(*map(set, [hdi.country, pop.country, gni.country, gender_equality.country, 
                                                       emissions.country, fundrights.country, murders.country, press_freedom.country])))
print(countries_in_common)
```

```python
# We print the shape of our dataframes to check a transformation later:
df_list = [hdi, pop, gni, gender_equality, emissions, fundrights, murders, press_freedom]
[i.shape for i in df_list]
```

```python
# As we can see at the dataframes description, the years in common are the following:
years_in_commmon = '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009','2012', '2013', '2014'
```

```python
# We'll use this function in our transformation:
def clean_dataframes (df):
    
    #We subset the dataframe by the countries all datasets have in common:
    df = df[df['country'].isin(countries_in_common)]
    
    #We subset the dataframe by years in common (2001-2014), 2010 and 2011 excepcions:
    df = df[['country','2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009','2012', '2013', '2014']]
    
    return df
```

```python
df_list = [df.pipe(clean_dataframes) for df in df_list]
```

```python
hdi = df_list[0]
pop = df_list[1]
gni = df_list[2]
gender_equality = df_list[3]
emissions = df_list[4]
fundrights = df_list[5]
murders = df_list[6]
press_freedom = df_list[7]
```

```python
#We test the adjustment by checking the shapes:
indices= [hdi, pop, gni, gender_equality, emissions, fundrights, murders, press_freedom]
[i.shape for i in indices]
```

# Data normalization:
We want to normalize all our variables so the range of the values is consistent, this normalization can make statistical analysis easier down the road. We want to transform all values into the range 0-1, the same as our reference indicator (hdi).

##Â Method One: Normalization dividing by 100:

```python
# The next dataframes have their values in range 0-100, we just divide by 100 then.
ls =[gender_equality, fundrights, press_freedom]
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
for i in ls:
    for c in [c for c in i.columns if i[c].dtype in numerics]:
        i[c] = i[c].div(100).round(2)
```

##Â Method Two: Normalization type Min-Max:

```python
# We normalize next dfs by subtracting the MIN and dividing by MAX minus MIN:
ls =[gni, emissions, murders]
```

```python
for i in ls:
    for c in [c for c in i.columns if i[c].dtype in numerics]:
        i[c] = ((i[c]-i[c].min())/(i[c].max()-i[c].min())).round(2)
```

# Data Analysis and Visualization:

At this point, we will analyze the correlations between indicators and we'll see which are the best candidates to add them to the HDI, thus increasing its dimensions.

```python
# Correlation dataframe:
corr = df.corr()
corr
```

```python
# With this interactive graph we can plot each correlation: try hdi-gni.
@interact
def correlations(column1=list(df.select_dtypes('number').columns),
                 column2=list(df.select_dtypes('number').columns)):
    print(f"Correlation: {df[column1].corr(df[column2])}")
    sns.jointplot(column1, column2, data=df, color='darkgreen')
```

```python
df.corrwith(df['hdi']).sort_values()
```

We can say most correlated indicators are gni, fundrights, gender_equality and murders (-).

```python
# We plot a correlation matrix:
mask = np.triu(np.ones_like(corr, dtype=np.bool))
cmap = sns.diverging_palette(275, 150, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=2, annot=True)
```

Since emissions and murders are negative indicators, we can adjust them to be positive so we can generate an index easily.

```python
df['emissions_adjusted'] = df.apply(lambda row: 1 - row.emissions, axis = 1)
df['murders_adjusted'] = df.apply(lambda row: 1 - row.murders, axis = 1)

#drop unadjusted columns:
df = df.drop(['emissions', 'murders'], axis=1)
```

```python
df.head(2)
```

To generate our new indicator on the HDI, two things must be taken into consideration: 

1. Based on the correlations between variables, we must discard the GNI and the fundrights indicators as they both have a high correlation with the HDI, also in the case of the fundrights, it is also highly correlated with gender_equality, so it is better to stick with gender_equality which has a lower correlation with the HDI.


2. We will weigh all the variables with the same weight (one seventh), taking into account that the HDI itself is made up of 3 variables, we will give it a weight of 3/7.

```python
# We generate the new dimensioned IDH called global_hdi:
df['global_hdi'] = df.apply(lambda row: (3*row.hdi/7)+(row.gender_equality/7)+(row.press_freedom/7)+(row.emissions_adjusted/7)+(row.murders_adjusted/7),
                            axis= 1)
df.head()
```

```python
# We observe correlation with hdi:
df = df[['country','year','hdi','global_hdi']]
corr = df.corr()
corr
```
